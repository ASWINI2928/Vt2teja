**`time`**

Only keep the first 16 characters of the `time` column, and change the column type to datetime.

```python
edf['time'] = pd.to_datetime(edf['time'], infer_datetime_format = True, utc = True)
edf['time'] = [str(x) for x in edf['time']]
edf['time'] = [x[:16] for x in edf['time']]
edf['time'] = pd.to_datetime(edf['time'], infer_datetime_format = True)
```

However, the timezone data is lost upon doing this and also doesn't take into account changes due to daylight saving time (happening in last week of October in all years), as shown below and in a quick [Google search](https://www.google.com/search?q=when+is+daylight+saving+time+spain&rlz=1C1EJFC_enMM843MM843&oq=when+is+daylight+saving+time+spain&aqs=chrome..69i57j0.8537j1j7&sourceid=chrome&ie=UTF-8).

```python
edf['time'].value_counts()
```

Look at these repetitions.

```python
repeats = edf['time'].value_counts().index[:4]
```

```python
edf[edf['time'].isin(repeats)]
```

Only `time` is repeated. Other columns look like they are genuinely different between the two repeated hours. This can be rectified by using `date_range` and set it as the time.

```python
edf['time'] = pd.date_range(start = '2015-01-01 00:00:00', end = '2018-12-31 23:00:00', freq = 'H')
```

```python
edf['time'].value_counts()
```

```python
edf[edf['time'].isin(repeats)]
```

Remove columns that have a lot of zeros in them, i.e. columns with very low or zero variance.

```python
std = edf.std()
std = std[std < 0.1]
std
```

```python
edf.drop(std.index, axis = 1, inplace = True)
```

```python
null = edf.isnull().sum()
null = null[null != 0]
null
```

Perform linear interpolation on the columns with missing data.

```python
for col in null.index:
    
    edf[col].interpolate(method = 'linear', axis = 0, inplace = True)
```

```python
edf.isnull().sum().sum()
```

```python
edf.reset_index(inplace = True)
```

```python
old_names = edf.columns
new_names = edf.columns.str.replace(' ','_')
```
